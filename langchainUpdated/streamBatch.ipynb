```python
##here we use the streaming and batching for generating the output not use the invoke function
import os
from dotenv import load_dotenv
load_dotenv()
gemini_api_key=os.getenv("GEMINI_API_KEY")
from langchain.chat_models import init_chat_model
model=init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
model.stream("what is ai in 200 words")
```

```python
for chunk in model.stream("what is ai  words"):
    print(chunk.text)
```

```python
response=model.batch(
    [
        "what is ai",
        "what is data science",
        "what is quantum computing"
    ],
    config={"max_concurrency":2}
)
for r in response:
    print(r.content.split("\n"))
```
