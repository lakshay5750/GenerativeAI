```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
```

```python
groq=ChatGroq(model="openai/gpt-oss-120b",groq_api_key=groq_api_key)
```

```python
groq
```

```python
from langchain_core.messages import SystemMessage,HumanMessage

messages=[
    SystemMessage("Convert the english language into hindi language"),
    HumanMessage("How are you ")
]
```

```python
result=groq.invoke(messages)
```

```python
groq.invoke(messages).response_metadata
```

```python
##we can use the output parser
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(result)
```

```python
chain=groq|parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate
generic_template="Convert this into following {language}:"
prompt=ChatPromptTemplate([
    ('system',generic_template),
    ('user',"{text}")
])
```

```python
prompt.invoke({"language":"hindi","text":"i am going to london"})
```

```python
chain=prompt|groq|parser
```

```python
chain.invoke({"language":"hindi","text":"i am going to london"})
```
