```python
from langchain_groq import ChatGroq
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.messages import HumanMessage,AIMessage
groq=ChatGroq(model="openai/gpt-oss-120b",groq_api_key=groq_api_key)
```

```python
groq.invoke([HumanMessage(content="Hii I am lakshya varshney. I am senior Data scientist")])
```

```python
groq.invoke([
HumanMessage(content="Hii I am lakshya varshney. I am senior Data scientist"),
AIMessage(content="Hello Lakshya! 👋 Great to meet a senior data scientist. How can I assist you today?"),
HumanMessage(content="what is my name what i do or proffesion of mine"),
HumanMessage(content="generative ai is very trending thing in today world"),
HumanMessage(content="machine learning is also growing"),
HumanMessage(content="data science is very big field"),
HumanMessage(content="what is my name")
])
```

```python
##memory
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
store={}
def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
```

```python
with_message_history=RunnableWithMessageHistory(groq,get_session_history)
```

```python
config={"configurable":{"session_id":"chat1"}}
```

```python
with_message_history.invoke(
    [HumanMessage(content="Hii I am lakshya varshney. I am senior Data scientist")],
    config=config
)
```

```python
with_message_history.invoke([
    HumanMessage("what is my name")
],config=config)
```
